Q2 Show bargraph of departure time and arrival time 

```python
data['departure_time'].value_counts()
```

```python
data['arrival_time'].value_counts()
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
# Get value counts
dep_counts = data['departure_time'].value_counts()
# Create bar chart
plt.bar(dep_counts.index, dep_counts.values)
# Add labels
plt.title('Departure Time Distribution')
plt.xlabel('Departure Time')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(16,4))
#for arrival time
plt.subplot(1,2,2)
arrival_counts = data['arrival_time'].value_counts()
# Create bar chart
plt.bar(arrival_counts.index, arrival_counts.values)
# Add labels
plt.title('Arrival Time Distribution')
plt.xlabel('Arrival Time')
plt.ylabel('Count')
plt.show()
```

Show bar graph representing the source city and destination city 

```python
data['source_city'].value_counts()
```

```python
data['destination_city'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))

# First subplot: Source Cities
plt.subplot(1, 2, 1)
source_counts = data['source_city'].value_counts()
plt.barh(source_counts.index, source_counts.values, color='skyblue')  # choose a single color or list of valid colors
plt.title("Source Cities with No. of Flights")
plt.ylabel("Cities")
plt.xlabel("No. of Flights")

# Second subplot: Destination Cities
plt.subplot(1, 2, 2)
dest_counts = data['destination_city'].value_counts()
plt.barh(dest_counts.index, dest_counts.values, color='lightgreen')  # valid color
plt.title("Destination Cities with No. of Flights")
plt.ylabel("Cities")
plt.xlabel("No. of Flights")

plt.tight_layout()  # adjusts spacing between subplots
plt.show()
```

Q5 Does ticket price change based on the departure time and arrival time

```python
#checking the mean ticket price based on the departure time
data.groupby('departure_time')['price'].mean()
```

```python
#checking the mean ticket price based on the arrival time
data.groupby('arrival_time')['price'].mean()
```

```python
sns.catplot(x='departure_time',y='price',kind='bar',data=data)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(
    x='arrival_time',
    y='price',
    kind='line',             # now valid
    data=data,
    col='departure_time',    # creates subplots for each departure time
    marker='o'               # adds points on the line
)

plt.show()
```
